```python
# Introduction to the data

import pandas as pd

# Read file
all_ages = pd.read_csv("all-ages.csv")
recent_grads = pd.read_csv("recent-grads.csv")

all_ages.head(5)
recent_grads.head(5)
```

```python
# Summarizing Major Categories

def calculate_major_cat_total(df):
    # Unique values in Major_category column
    cats = df["Major_category"].unique()
    # Dictionnaire à remplir
    counts_dict = dict()
    
    for current_major in cats:
        # Tous les rows avec le major courant
        is_current_major = df["Major_category"] == current_major
        major_df = df[is_current_major]
        # Calculer le total d'étudiants pour le major courant (sur toutes les rows)
        total = major_df["Total"].sum()
        # Ajouter au dico {major, total}
        counts_dict[current_major] = total
    return counts_dict
        
        
aa_cat_counts = calculate_major_cat_total(all_ages)
rg_cat_counts = calculate_major_cat_total(recent_grads)

print(aa_cat_counts)
print("\n")
print(rg_cat_counts)
```

```python
# Low-Wage Job Rates

# Récupération des 2 colonnes
low_wage_col = recent_grads["Low_wage_jobs"]
total_student_col = recent_grads["Total"]

sum_low_wage = low_wage_col.sum()
total_student = total_student_col.sum()

low_wage_percent = float(sum_low_wage) / float(total_student)
print(low_wage_percent)
```

```python
# Comparing Data Sets

# All majors, common to both DataFrames
majors = recent_grads['Major'].unique()
rg_lower_count = 0

for current_major in majors:
    # All rows with the current major
    major_aa = all_ages[all_ages["Major"] == current_major]
    major_rg = recent_grads[recent_grads["Major"] == current_major]
    
    # Unemployment_rate
    enemployment_rate_aa = major_aa["Unemployment_rate"]
    enemployment_rate_rg = major_rg["Unemployment_rate"]
    
    enemployment_rate_aa_min = enemployment_rate_aa.min()
    enemployment_rate_rg_min = enemployment_rate_rg.min()
    
    if enemployment_rate_rg_min < enemployment_rate_aa_min:
        rg_lower_count += 1

print("Il y a " + str(rg_lower_count) + " majeures où le taux de chômage est + faible dans les résultats récents que sur les résultats globaux")
    
    
```
